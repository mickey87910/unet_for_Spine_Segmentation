```python
import os
import glob
from PIL import Image
image_files = glob.glob('spine_img/train/image/*.bmp')
mask_files = glob.glob('spine_img/train/label/*.png')
test_image_files = glob.glob('spine_img/test/*.bmp')
```

```python
for file in mask_files:
    img = Image.open(file)
    new_image = Image.new("RGBA", img.size, "WHITE") # Create a white rgba background
    new_image.paste(img, (0, 0), img)
    new_image.convert('RGB')
    new_image.save(file)
    os.rename(file, '{}.png'.format(file.split('-1')[0]))
```

```python
for file in image_files:
    img = Image.open(file)
    img.save(file[:-4]+".png","png")
    os.remove(file)
    #os.rename(file, '{}.png'.format(file.split('.')[0]))
```

```python
for file in test_image_files:
    img = Image.open(file)
    img = img.resize((256,256))
    img.save(file[:-4]+".png","png")
    os.remove(file)
```
